    <li>Write a function that converts a number into US standard money formatting (every 3 digits contain a comma and demical separates cents from dollar). That is, design a function that takes a number like $123456789.10$ and outputs $ \$123,456,789.10$ </li>

```python
# 1

def underscore(contlist):
    string = contlist[0]
    for i in contlist[1:]:
        string = string+'_'+i
    return string
```

```python
underscore(['this', 'is', 'a', 'list'])
```

```python
# 2

def factorial(n):
    prod = 1
    for i in range(1,n+1):
        prod *= i
    return prod
```

```python
factorial(3)
```

```python
# 3

def fibonacci(n):
    fib1 = 0
    fib2 = 1
    for i in range(n):
        fib1, fib2 = fib2, fib1+fib2
    return fib2
```

```python
for i in range(5):
    print(fibonacci(i))
```

```python
# 4

def binary(n):
    if n <= 0:
        return "0"
    binarie = ""
    while n > 0:
        remainder = n % 2               
        binarie += str(remainder)
        n //= 2                       
    return binarie[::-1]
```

```python
binary(18)
```

```python
# 5

def money(string):
    string = string.replace('$', '')
    string = string.replace(',', '')
    return string
```

```python
# 7

def summation(contlist):
    summ = 0
    for i in contlist:
        summ += i
    return summ
```

```python
summation([1,2,3])
```

```python
# 8

def minimum(contlist):
    minimum = contlist[0]
    for i in contlist:
        if i<minimum:
            minimum = i
    return minimum
```

```python
minimum([1,2,3])
```

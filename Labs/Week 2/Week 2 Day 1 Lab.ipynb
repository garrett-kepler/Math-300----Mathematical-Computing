Consider the function from Friday's quiz:
```python
def squares(maximum):

    container = []

    for i in range(maximum):
        container.append(i)

    container2 = []

    for j in container:
        container2.append(j**2)

    container3 = 0

    for k in container2:
        container3 += k

    return container3
```

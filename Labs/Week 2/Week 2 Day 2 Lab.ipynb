### Problem 4

The derivative of a function \(f(x)\) at a point \(x\) is defined as the limit

$
f'(x)=\lim_{h\to0}\frac{f(x+h)-f(x)}{h}
$

provided the limit exists.

Define a function $g(h)$ that computes the quotient above for a given function $f$ near a point $x$. Then, vectorize $g$. 


Using `np.linspace()`, let $f(x)=sin(x)$ and compute $g$ at $10^{-1}, 10^{-2}, 10^{-3}, 10^{-4}$. This will be an estimate for the derivative of $f$. Compare to the actual derivative.

````python
import math
````

````python
x = np.pi # some point x

def g(h):                  
    return (f(x+h)-f(x))/h  # quotient for derivative

G = np.vectorize(g)   # vectorize g
def f(x):
    return math.sin(x) # f of choice

h = np.linspace(-1, -4, 4)   # -1, -2, -3, -4
h = 10**h                   # 10^-1, 10^-2, 10^-3, 10^-4

print(G(h))                 # evalutate the vectorized version of g at h values
print(math.cos(np.pi))      # actual derivative
````

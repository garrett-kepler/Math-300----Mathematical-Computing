### Problem 4

Recall:
- `np.linspace(start, stop, slice_count)` produces `slice_count` amount of numbers between `start` and `stop`
- `np.sin(x)` calculates $\sin(x)$ where `x` is a single number or numpy array (which `np.linspace` produces)
- `np.pi` produces the irrational number $\pi\approx 3.14159$

Using these functions, 
1. make a list `xdata` that contains 100 evenly spaced points between 0 and $2\pi$
2. make a list `ysindata` that contains $\sin(x)$ for each $x$ in `xdata`
3. use `plt.plot(xdata, ysindata)` to plot the data

````python
xdata = np.linspace(0, 2*np.pi, 100)
ysindata = np.sin(xdata)
plt.plot(xdata, ysindata)
````

### Problem 5
Using the same `xdata` from above, make a list `ycosdata` that contains $\cos(x)$ for each $x$ in `xdata`.

We can plot `xdata` against both `ycosdata` and `ysindata` simultaneously. To do so, use `plt.plot(xdata, ysindata)` and `plt.plot(xdata, ycosdata)` in the same cell.

````python
ycosdata = np.cos(xdata)
````

````python
plt.plot(xdata, ysindata) 
plt.plot(xdata, ycosdata)
````

So far, these just look like two curves. To differentiate between the two curves, we can use the `label` attribute and a legend. 

All in one cell:
1. Instead of `plt.plot(xdata, ysindata)`, use `plt.plot(xdata, ysindata, label='Sine')`. Likewise with the cosine data. 
2. Use `plt.legend()` to plot the legend.

````python
plt.plot(xdata, ysindata, label='Sine')
plt.plot(xdata, ycosdata, label='Cosine')
plt.legend()
````

Now we know which curve is which! But, we don't have axis labels or a title. Add these in the following cell.

````python
plt.xlabel('x value')
plt.ylabel('Function value')
plt.title('Cosine vs Sine')
plt.plot(xdata, ysindata, label='Sine')
plt.plot(xdata, ycosdata, label='Cosine')
plt.legend()
````

### Problem 6
Consider the following data set that contains values between 0 and 7 (inclusive):

```python
data = [4, 2, 4, 5, 0, 2, 4, 4, 3, 6, 1, 4, 5, 3, 1, 2, 3, 3, 4, 6, 4, 5,
       2, 6, 5, 2, 3, 3, 4, 4, 2, 1, 4, 3, 2, 3, 6, 7, 3, 3, 3, 4, 2, 4,
       3, 2, 2, 2, 4, 3, 2, 3, 4, 2, 5, 5, 3, 3, 5, 3, 2, 6, 2, 2, 4, 4,
       2, 3, 4, 5, 2, 3, 4, 6, 2, 5, 2, 4, 5, 1, 5, 3, 3, 5, 0, 2, 3, 5,
       3, 2, 2, 5, 7, 5, 3, 3, 4, 2, 4, 4]
```
1. Write a function that will take in `data` and outputs a list `counts` containing the number of each value in `data`. (Hint: `data.count(i)` gives you the count of `i` in `data`)


````python
data = [4, 2, 4, 5, 0, 2, 4, 4, 3, 6, 1, 4, 5, 3, 1, 2, 3, 3, 4, 6, 4, 5,
       2, 6, 5, 2, 3, 3, 4, 4, 2, 1, 4, 3, 2, 3, 6, 7, 3, 3, 3, 4, 2, 4,
       3, 2, 2, 2, 4, 3, 2, 3, 4, 2, 5, 5, 3, 3, 5, 3, 2, 6, 2, 2, 4, 4,
       2, 3, 4, 5, 2, 3, 4, 6, 2, 5, 2, 4, 5, 1, 5, 3, 3, 5, 0, 2, 3, 5,
       3, 2, 2, 5, 7, 5, 3, 3, 4, 2, 4, 4]
````

````python
def count(dataset):
    counts = []
    for i in set(dataset):
        counts.append(dataset.count(i))
    return counts
````

````python
count(data)
````

2. Use `plt.pie(counts)` to produce a pie chart of `data`.

````python
counts = count(data)
plt.pie(counts)
````

3. Define `labels = [0, 1, 2, 3, 4, 5, 6, 7]` and use `plt.pie(counts, labels = labels, autopct='%1.1f%%')` to produce more informative plot

````python
labels = [0, 1, 2, 3, 4, 5, 6, 7]
plt.pie(counts, labels = labels, autopct='%1.1f%%')
````

4. Add an informative title and produce the new plot in the cell below.

````python
plt.title("This is a Pie Chart")
plt.pie(counts, labels = labels, autopct='%1.1f%%')
````

### Problem 7
Using the same data set as above, use `plt.hist()` to produce the histogram of the data set. Add a title and labels.

````python
plt.hist(counts)
plt.title('This is a Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
````

### Problem 1
Suppose $f(x)=x^2+6x+10$

1. Plot the function to visualize the minimum
2. Find the minimum using `scipy`

````python
import scipy as sp
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(-10,10)
plt.plot(x, x**2+6*x+10)
````

````python
from scipy.optimize import minimize
````

````python
x0 = 0
def f(x):
    return x**2+6*x+10
````

````python
minimize(f, 
         x0=x0)
````

````python
f(-3.00000003)
````

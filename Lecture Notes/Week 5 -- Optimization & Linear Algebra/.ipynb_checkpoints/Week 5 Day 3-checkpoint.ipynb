````python
# preliminaries 
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize 
````

# Optimization in Machine Learning

Most of machine learning is simply solving optimization problems. For example, we discussed linear regression last week. Formally, we often implement linear regression to minimize the *squared error*.

We want a $\hat y = mx+b$ that minimizes
$$ SSE=\sum_{i=1}^{n}(y_i-\hat y_i)^2$$
The goal is simple: Find the values of $m$ and $b$ that make the SSE as small as possible.

For example, let
```python
x = np.array([1,2,3,4,5])

y = np.array([2,5,5,8,9])
```
Suppose $y=mx+b$. We wish to find $m$ and $b$ that minimize squared error. So, we can define an objective function to minimize:


````python
x = np.array([1,2,3,4,5])
y = np.array([2,5,5,8,9])
def loss(params):
    a,b = params
    prediction = a*x+b
    return np.sum((y-prediction)**2)

result = minimize(loss,[1,1])

print(result.x)
````

````python
a,b = result.x
plt.figure(figsize=(7,5))
plt.scatter(x,y)
plt.plot(x,a*x+b)
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Linear Regression Plot')
plt.show()
````

Always good to check different starting values!

````python
for guess in [[-1,-1],[1,2],[0,4],[0,8]]:
    result = minimize(loss,
                      guess)

    print(result.x)
````



This is the basis of most machine learning algorithms are optimization problems.


| Model | What Is Minimized? |
|--------|--------------------|
| Linear Regression | Sum of Squared Errors |
| Logistic Regression | Log Loss |
| Neural Networks | Cross-Entropy Loss |
| Ridge Regression | Squared Error + Penalty |
| Lasso Regression | Squared Error + L1 Penalty |

Although these models appear different, they all share the same idea:

1. Define a loss (objective) function.
2. Measure how well the model fits the data.
3. Adjust the model parameters to reduce the loss.
4. Stop when no better solution can be found.

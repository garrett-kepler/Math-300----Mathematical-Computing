### Vectors
Recalling your experiences in Math 220, we typically start linear algebra with vector operations. In Python, typically, vectors are one-dimensional numpy arrays:

```python
x = np.array([1,2,3,4])
y = np.array([-1,0,2,0])
```

```python
x
```

```python
print(x + y)
print(x - y)
```

Remember why numpy arrays are so nice: they act like numbers! We can add/subtract vectors:

We can scale vectors:

```python
2 * x
```

We can also take the dot product of vectors:

```python
print(np.dot(x, y))
print(x @ y)
```

```python
np.outer(x,y)
```

The norm a vector can be easily computed so that we can normalize:

```python
norm = np.linalg.norm(x)

normalizedx = x/norm

print(norm)
print(normalizedx)
print(np.linalg.norm(normalizedx))
```

### Matrices
We can likewise perform operations on matrices!

To define a matrix, we can do two things:

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([1,2,3,4]).reshape(2,2)

print(A)
print(B)
```

If you need to know a matrix's size, simply use the `.shape` attribute:

```python
A.shape
```

Still, these are numpy arrays. Basic operations work exactly as we would like:

```python
print(A + B)
print(3 * A)
```

To perform matrix multiplication, we simply use `np.dot()` or `@`:

```python
print(np.dot(A, B))
print(A @ B)
```

```python
print(A*B) # elementwise multiplication
```

We can compute the transpose in a couple ways:

```python
print(A.T )
print(np.transpose(A))
```

For the typical properties of matrices we like, numpy has them built in!

```python
# inverse

print(np.linalg.inv(A))

# determinant

print(np.linalg.det(A))

# rank

print(np.linalg.matrix_rank(A))

# trace

print(np.trace(A))

# eigenvalues/eigenvectors

print(np.linalg.eig(A))
```

```python
np.linalg.eig(A)[1]
```

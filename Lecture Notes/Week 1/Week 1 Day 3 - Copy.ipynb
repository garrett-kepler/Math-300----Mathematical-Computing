## Variables 
Last class, we implicitly introduced the concept of a "variable" in Python. To <i>assign</i> a variable `x`, we simply use `x = whatever_we_want`. For example,

````python
x = 5
y = ['this is a string: '+str(3.14), 'this is a number: '+str(int('4'))]

print(x)
print(y)
````

After assignment, Python will understand any operation on the variable as an operation on the actual object itself. That is, once we have a variable, we can perform operations on the <i>representation</i> of the object. For example,

````python
x**2 + x - 2
````

````python
x**2 + x - 2 == 5**2 + 5 - 2
````

We can also reassign at will. But, <i>whatever the last running assignment is for a variable, that will be the the object Python refers to</i>. For example,

````python
x = 5   # initially x is assigned 5
x = 4   # by reassigning x to be 4, x is 4 for every line of code after this
print(x)
````

## Conditional Statements
One thing we will cover today is how to use conditional statements in Python. By "statement", we mean a line of code that does not transform any object (for instance `x = x + 2`), but determines the truth value (boolean value) of a phrase (for instance `x > 2`). For example:

````python
x > 3 # x = 4
````

A <i>conditional statement</i> utilizes `if`, `else`, and `elif` to determine whether or not to perform an operation. A straightforward example:

````python
x = 4
if x > 3:    # if variable x is greater than 3
    x **= 2  # then reassign x as x squared (x = x**2)
x
````

What Python reads this as is "if A is true, then perform B". If we want to have an alternative operation in the case that A is not true, we can utilize the `else` statement:

````python
x = float(input('what would you like x to be? ')) # a simple input to demonstrate if else statement

if x > 3:    # if variable x is greater than 3
    x **= 2  # then reassign x as x squared (x = x**2)

else:           # if variable x is less than or equal to 3
    x += 2  # then reassign x as x + 2 (x = x + 2) 
    
print(x)
````

All `if` statements are <i>independent</i>. That is, multiple statements can be true at the same time. 

````python
x = 10

if x < 100:
    x += 1   # x is now 11
if x < 1000: # since x is 11, this still applies
    x += 1
    
print(x)
````

The `elif` statement allows you to make more conditions than a simple `if` `else` block. `elif` meaning "else if" follows an `if` statement.

````python
x = 10

if x < 100:    # this applies to x
    x += 1     # x is now 11
elif x < 1000: # since x already went through the above if statement,
    x += 1     #  this doesn't apply
    
print(x)
````

These can be useful if you want to do piecewise functions or anything involving multiple conditions. For example, the function

$\begin{align*}
f(x)&=\begin{cases}
        x+2~~\text{if } 2\geq x\geq 1\\
        x-2~~\text{if } 3\geq x\geq 2\\
        x^2~~\text{if } x\geq 3\\
        2x~~\text{otherwise}
        \end{cases}
        \end{align*}$
        
could be executed using the following cell.

````python
x = float(input('what would you like x to be? ')) 

if 2 >= x >= 1:
    x += 2
    
elif 3 >= x >= 2:
    x -= 2  
    
elif x >= 3:    
    x **= 2  
    
else:          
    x *= 2  
    
print(x)
````

## Conditional Statement Operations
Some classic useful operations you may need when using conditional statements are the following:

````python
print(x)

# Less than
print(x < 10)

# less than or equal to 
print(x <= 10)

# greater than
print(x > 10)

# greater than or equal to
print(x >= 10)

# equal
print(x == 10)

# not equal
print(x != 10)

# object equality
print(x is 10)

# not object equality
print(x is not 10)
````

You can use these to form multiple pairwise comparisons as well:

````python
x = 11
12 > x > 10
````

````python
# note that these are pairwise comparisons

12 > x < 10
````

This also works for other object types as well.

````python
print(str(10) == '10')
````

The difference between `==` and `is` lies in what is being asked. The comparison `==` is checking whether the values of objects are the same. Meanwhile, the comparison `is` checks whether the objects themselves are the same as well. For example,

````python
x = 1   # integer
y = 1.0 # float

print(x == y) # the value is the same (it's 1)
print(x is y) # the objects are not the same because they are different types
````

These operations also extend to lists as well. The `>`, `<`, `>=`, `<=`, `==` and `!=` will make entrywise comparisons between two lists. That is, for each `>` comparison, Python asks "is `list1[i] > list2[i]` true?".

````python
print([1,2,3] <= [1,2,3])
print([1,2,3] >= [1,3,2])
````

Though, this can be counterintuitive at times.

````python
print([1,5,5] < [2,0,0])
````

Likewise, with the other operations `is` and `is not`. But, again, these ask if the <i>objects themselves</i> are also the same.

````python
x = ['square root of pi', 1.772]
y = ['square root of pi', 1.772]

print(x == y) # the values inside the lists are the same
print(x is y) # these objects are two distinct lists so not the same object
print(x is ['square root of pi', 1.772]) # these objects are two distinct lists so not the same object
print(x is x) # these are the same object
````

### For-loops
For-loops are essential to Python programming. "For-loop" because they begin with ``for`` and loop through some operation. They allow us to iterate through an object and perform an operation for each element of the object. For example, we can operate on the object itself:

````python
contlist = [1,2,3]
contlist2 = []
contlist3 = []
for i in contlist: # for each element in contlist
    contlist2.append(i+1) # add one to that element and append it to contlist2
    contlist3.append(i**2)
print([contlist, contlist2, contlist3])
````

Or, even operate on an entirely different object:

````python
x = 1
for i in [1,2,3,4,5,6,7,8,9]:
    if i>3:
        x = x*.5
        print(x)
````

For-loops are generically formatted as the above two examples. We have some object we iterate over (called an <i>iterable</i>) and assign a variable that will iterate through this iterable object.

A for-loop format is:
```python

    for whatever_variable_you_want in whatever_iterable_you_want:
        some_lines_of_code_that_execute_something
```
        
Importantly, like we have discussed at the beginning, the last assignment of a variable is what Python remembers. That is, whatever variable you decide to use in this loop will be assigned the last value it has obtained.

````python
for i in range(15):
    if i < 100000:
        'just look at the output, it will be the last number i is defined to be'
print(i)
````

### Nested For-Loops
One important feature of for-loops follows the same logic as conditional statements: we can do more than one. For-loops can be compounded (what we call <i>nested</i>). We can have a for-loop inside of a for-loop inside of a for-loop and so on.

````python
count = 0
for i in range(5):
    for j in range(5):
        count += 1
        print('i am step '+str(count))
````

This can be incredibly useful. Take for example a problem from last lab:

Find all integer solutions $(a,b)$ satisfying

$\begin{align*}
a+b=25\\
a\cdot b= 144
\end{align*}$

We can solve this with a for-loop:

````python
for a in range(25):
    b = 25 - a
    if a*b == 144:
        print([a,b])
````

We can also solve this with a nested for-loop:

````python
for a in range(25):
    for b in range(25):
        if (a*b == 144) & (a+b==25):
            print([a,b])
````

We can even solve this with four for-loops and three conditional statements:

````python
for a in range(25):
    for b in range(25):
        if a+b == 25:
            for c in range(25):
                for d in range(25):
                    if c*d == 144:
                        if [a,b] == [c,d]:
                            print([a,b])
````

I hope you can see the difference between these solutions. Mainly, I hope you can see that the last method is absolutely more complicated than the first two. This is the key: <i>coding is a tool that we should use to our advantage</i>. Correct convoluted code will always work. If that is the only method we have that works, stick with it! If there is a simple way, maybe that's better. At the end of the day, solving problems is the goal. But, communicating reproducable code is absolutely second on the list.

The last note on this difference: time.

````python
import time
````

````python
start = time.time()
for a in range(25):
    b = 25 - a
    if a*b == 144:
        print('found a solution')
stop = time.time()
print('Method 1 took '+str(stop-start)+' seconds')

start = time.time()
for a in range(25):
    for b in range(25):
        if (a*b == 144) & (a+b==25):
            print('found a solution')
stop = time.time()
print('Method 2 took '+str(stop-start)+' seconds')

start = time.time()
for a in range(25):
    for b in range(25):
            for c in range(25):
                for d in range(25):
                    if c*d == 144:
                        if a+b == 25:
                            if [a,b] == [c,d]:
                                print('found a solution')
stop = time.time()
print('Method 3 took '+str(stop-start)+' seconds')
````

Method 1 in this case took $<10^{-16}$ seconds to finish. Method 3 in this case took $\approx 10^{-1}$ seconds to finish. This is more than 15 powers of 10.

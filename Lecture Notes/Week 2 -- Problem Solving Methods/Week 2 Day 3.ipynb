# The `math` Library

The `math` library is included with Python and provides functions for common mathematical operations on <i>scalars</i>. 

```python
import math
```

```python
print(math.sqrt(25))         # square root
print(math.sin(math.pi/2))   # sine
print(math.cos(0))           # cosine
print(math.log(10))          # natural logarithm
print(math.exp(2))           # exponential (e^x)
print(math.factorial(5))     # factorial
```

# The SymPy Library

SymPy performs what we call <i>symbolic mathematics</i> rather than solving numerically like in numpy. To do this, we reform arithmetic in terms of a "sympy symbol". This object takes much longer to process and deal with than floats, so be careful!


```python
import sympy as sp
```

To create a variable in sympy, we must tell Python to create a sympy symbol object:

```python
x = sp.symbols('x')
```

Now, anytime the variable `x` is mentioned, Python knows to refer to the sympy-based arithmetic of the object. One basic operation in sympy is simplification:

```python
print(sp.simplify((x**2 - 1)/(x - 1)))
```

Also, expansion:

```python
print(sp.expand((x + 2)**5))
```

We can also factor polynomials:

```python
print(sp.factor(x**2 - 9))
```

And solve quadratics: 

```python
solution = sp.solve(x**2 - 4, x) # sp.solve( function, variable )
print(solution)
```

Or evaluate limits:

```python
print(sp.limit(sp.sin(x)/x, x, 0)) # sp.limit( function, variable, limit )
```

Even differentiate:

```python
f = x**3 + 2*x**2 - x
print(sp.diff(f, x))
```

And integrate:

```python
print(sp.integrate(x**2, x)) # indefinite integral format

print(sp.integrate(x**2, (x, 0, 4))) # definite integral (from 0 to 4)
```

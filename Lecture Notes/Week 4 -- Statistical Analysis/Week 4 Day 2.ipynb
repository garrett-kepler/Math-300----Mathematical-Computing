```python
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
from ipywidgets import interact, Text, FloatSlider
from IPython.display import display
from scipy.stats import beta, norm
from scipy import stats
```

### Basics Statistics Review

```python

def beta_dist(alpha=2.0, beta_param=2.0, sample_size=5000):
    sample = np.random.beta(alpha, beta_param, sample_size)
    mean = np.mean(sample)
    median = np.median(sample)
    counts, bins = np.histogram(sample, bins=50)
    mode_sample = (bins[np.argmax(counts)] + bins[np.argmax(counts)+1]) / 2

    variance = np.var(sample, ddof=1)
    std = np.std(sample, ddof=1)
    mad = np.mean(np.abs(sample - mean))
    
    if alpha > 1 and beta_param > 1:
        theoretical_mode = (alpha - 1)/(alpha + beta_param - 2)
    else:
        theoretical_mode = None

    x = np.linspace(0,1,500)
    pdf = beta.pdf(x, alpha, beta_param)

    plt.figure(figsize=(10,5))
    plt.plot(x, pdf, linewidth=3, label='Beta PDF')
    plt.fill_between(x, pdf, alpha=0.25)

    plt.axvline(mean, linestyle='--', linewidth=2,label=f"Mean = {mean:.3f}", color='red')
    plt.axvline(median, linestyle='-.', linewidth=2,label=f"Median = {median:.3f}", color='green')
    if theoretical_mode is not None:
        plt.axvline(theoretical_mode, linestyle=':',
                    linewidth=3,
                    label=f"Mode = {theoretical_mode:.3f}")

    plt.title(f"Beta Distribution (α={alpha:.1f}, β={beta_param:.1f})")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.xlim(0,1)
    plt.grid(alpha=0.3)
    plt.legend()

    plt.show()

    print("Sample Statistics")
    print("-----------------------------")
    print(f"Mean:                    {mean:.4f}")
    print(f"Median:                  {median:.4f}")
    print(f"Approximate Sample Mode: {mode_sample:.4f}")
    if theoretical_mode is not None:
        print(f"Theoretical Mode:        {theoretical_mode:.4f}")
    else:
        print("Theoretical Mode:        Undefined (α ≤ 1 or β ≤ 1)")
    print(f"Variance:               {variance:.4f}")
    print(f"Standard Deviation:     {std:.4f}")
    print(f"Mean Absolute Deviation:{mad:.4f}")

interact(
beta_dist,
alpha=FloatSlider(
    value=2,
    min=0.5,
    max=10,
    step=0.1,
    description='α'
),
    beta_param=FloatSlider(
        value=2,
        min=0.5,
        max=10,
        step=0.1,
        description='β'
    ),
    sample_size=(100,20000,100)
);
```

```python
def normal_dist(mean=0, sigma=1):
    sample = np.random.normal(mean, sigma, 5000)
    x = np.linspace(mean - 4*sigma, mean + 4*sigma, 500)
    y = norm.pdf(x, mean, sigma)
    
    sample_mean = np.mean(sample)
    sample_median = np.median(sample)
    counts, bins = np.histogram(sample, bins=50)
    
    mode = (bins[np.argmax(counts)] + bins[np.argmax(counts)+1])/2
    variance = np.var(sample, ddof=1)
    std = np.std(sample, ddof=1)
    mad = np.mean(np.abs(sample - sample_mean))

    plt.figure(figsize=(10,5))
    plt.plot(x, y, linewidth=3, label="Theoretical PDF")
    plt.hist(sample,bins=40,density=True,alpha=0.35,label="Random Sample")
    plt.axvline(sample_mean,linestyle="--",linewidth=2,label=f"Mean = {sample_mean:.2f}", color='red')
    plt.axvline(sample_median,linestyle="-.",linewidth=2,label=f"Median = {sample_median:.2f}", color='orange')
    plt.axvline(mode,linestyle=":",linewidth=3,label=f"Approx. Mode = {mode:.2f}", color='black')
    plt.title(f"Normal Distribution (μ={mean:.1f}, σ={sigma:.1f})")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.xlim(-1,1)
    plt.ylim(0,2)
    plt.grid(alpha=0.3)
    plt.legend()
    plt.show()

    print("Descriptive Statistics")
    print("------------------------------")
    print(f"Mean:                    {sample_mean:.4f}")
    print(f"Median:                  {sample_median:.4f}")
    print(f"Approximate Mode:        {mode:.4f}")
    print(f"Variance:               {variance:.4f}")
    print(f"Standard Deviation:     {std:.4f}")
    print(f"Mean Absolute Deviation:{mad:.4f}")

interact(normal_dist,mean=FloatSlider(value=0,min=-5,max=5,step=0.5,description='μ'),
    sigma=FloatSlider(
        value=1,
        min=0.2,
        max=1,
        step=0.05,
        description='σ'
    )
);
```

### Sequences

There a lot of objects we can iterate through. But, the typical ones we deal with are the <i>sequences</i>. A list is a type of sequence. A string is also another type of sequence. Both a list and a string are a collection of items where order matters. That is a sequence. For example,

````python
# list
contlist = [1,2,3,4]

# string
contstring = '1234'

# tuple
conttuple = (1, 2, 3, 4)
````

### Working with Sequences
Just like other object types, we can perform operations on sequences as well.

````python
# concatenating sequences with +

print([1, 2, 3] + [3, 4])
print('stri' + 'ng')
````

````python
# repeating sequence with *

print( (11,12,13) * 4 )
print('abcd' * 8)
````

````python
# determining number of elements with len()

print( len(contlist) )
print( len('12saf+-j') )
````

````python
# obtaining index of element (first occurence) with .index()

print( contlist.index(2) )
print( '1234'.index('4') )
````

````python
# counting number of occurences of element with .count()

print( (11, 12, 13, 11, 12, 13, 11, 12, 13, 11, 12, 13).count(11) )
````

````python
# checking if object is an element of sequence with in

contlist = [1,2,3,4]
contstring = 'the sky is blue'
print(1 in [1, 2, 3, 4])
print('blue' not in contstring)
````

Remark: ``` in``` and `not in` will check if there is an <i>element</i> of the sequence that matches. For example, `in` will not let you check for multiple elements at a time:

````python
[1,2] in [1,2,3,4]
````

Because `in` thinks of `[1,2]` as one object:

````python
[1,2] in [[1,2], 3, 4]
````

### Indexing and Slicing

Like we learned a couple days ago, we can access items of a list via an index: `list_name[desired_index]`.

````python
contlist = [12,14,15,13]
contlist[2]
````

The same is true for all sequences:

````python
contstring = 'the sky is blue'
conttuple = (11, 12, 13, 11, 12, 13, 11)

print(contstring[5])
print(conttuple[3])
````

We can take a subset of indices to obtain a <i>slice</i> as well. Writing `list_name[start:stop:step]` will take the element `list_name[start]` first, then `list_name[start + step]`, then `list_name[start + 2*step]`, and so on until it runs out of elements to select. 

````python
contlist = [1,2,3,4,5,6,7,8]

contlist[1:7:3] # starting at 1 and ending at 7 skipping everything 3
````

If `start` is not included, Python defaults to the first index.
If `stop` is not included, Python defaults to the last index.
If `step` is not included, Python defaults to step size 1.

````python
contstring = 'example'


print(contstring[:4:2]) #starting at 0, ending at 4, every 2nd element
print(contstring[3::3]) #starting at 3, ending at 6, every 3rd element
print(contstring[1:6:]) # starting at 1, ending at 6, every element
````

Note: the element `list_name[stop]` is excluded:

````python
contstring[0:6] # returns contstring[0] through contstring[5]
````

Negative values also work! This is reversed order. That is, if `seq` is a sequence of length `n`, then `seq[0] == seq[-n]`, `seq[1] == seq[-(n-1)]`, ..., and `seq[n-1] == seq[-1]`. For example,

````python
contstring = 'coding'
n = len(contstring)
count = 0
for i in contstring:
    print(i+' is at index '+str(count)+' or equivalently '+str(-(n-count)))
    count+=1
````

Negative values aren't just for indices. They work for step size as well:

````python
conttuple = (1,2,3,4,5,6,7,8,9)

conttuple[::-1]
````

Note: pick an out of bounds index will always raise an error, but an out of bounds slice will not. For example,

````python
conttuple[1000] # there is no element at the 1000th index
````

````python
conttuple[1:1000:2] # pick up all values that lie in between the start and stop
````

### Functions Introduction
Functions in Python work very similar to nice mathematical functions. They take inputs and produce outputs. We have actually encountered some built-in functions: `print()`, `len()`, `max()`, etc. What Python allows us to do is create our own functions.

The typical format for a function:

```python

    def function_name(input_1, input_2, ..., input_n):
        some block of code doing something involving the inputs
        return output_1, output_2, ...., output_m
```

`def` here named after "define". `def` tells Python to create a function object for you. `function_name()` can be any variable name you like. Once the function `function_name` has been executed and the operations described by the block code within `function_name` have finished, `return` will end running `function_name` and produce the outputs you specify (`output_1,...,output_m`).

Here is a simple example:

````python
def polynomial(x): # the function name will be `polynomial` with one input: x
    """ Since my variable is x, the code here should refer to x """
    
    if x < 0: 
        y = 1/x
        
    else:   # if not x < 0
        y = x**2+5
        
    return y # single output: y
````

Now that I have defined `polynomial()`, I can input values into it.

````python
polynomial(-5) # by definition of polynomial(), 5 > 0 so should be 30 
````

Functions don't even need inputs!

````python
def contfunction():
    print('you ran the function')
contfunction()
````

They will always output something though (even if it is only `None`)

Functions are senstive to syntax errors just like everything else in Python. Remember to have parantheses, colons, and indentation.

These are all bad:

```python

    def function(inputs)
        code
        return outputs
    
    def function(inputs):
    code
        return outputs
    
    def function(inputs):
        code
    return outputs

    def function:
        code
        return outputs
```

This is good:

```python

    def function(inputs): # def then name and inputs with parantheses and colon
        code  # indented block of code
        return outputs # return indented
```

One last one:

````python
def squaresum(sequence): # a function that takes a sequence as input and outputs the sum of squares in the sequence
    
    contsum = 0
    
    for i in sequence:
        contsum += i**2
    return contsum
````

#### Numbers
Just like in mathematics, there are different types of numbers. An "integer" and "complex number" in Python is exactly what it is in math. A "float" is any real-valued number (named after floating point number).  

When coding, working with these is nearly the same as working with a calculator. Here are some examples.

```python
# Addition
print(10 + 3)  

# Subtraction
print(10 - 3) 

# Multiplication
print(10 * 3)  

# Division
print(10 / 3) 

# Integer division (floor)
print(10 // 3) 

# Modulus 
print(10 % 3)  

# Exponentiation
print(10 ** 3)
```

```python
type(3.5)
```

Operations between two of the same type doesn't necessarily preserve that object type. 

```python
type(10/5) # Integer divided by integer gives float
```

```python
type(int(10/5))
```

An integer can be converted to a float exactly by using `float()`. But, converting a float to a integer by using `int()` forces rounding.

```python
print(float(4))
print(int(4.2))
```

#### Booleans
"Booleans" are simply binary values. In Python, there are two boolean-types: `True` and `False`. Indeed this is `True` and not `true`. Python is a case-sensitive programming language.

These are often useful to sanity check. For example, checking whether 10 is greater than 3:

```python
print(10 > 3)
```

We can even formalize the above statement using an `if` statement as follows.

```python
if 10 < 3:
    print(True)
```

We often like to do logical operations between statements. For this, we can use the phrases `and` or `or`. Alternatively, we can use `&` and `|` respectively. 

```python
print((10 > 3) & (10 > 5))
print((10 < 3) | (10 > 5))
```

Behind the screen, `True` and `False` are stored as 1 and 0. We can see this by checking whether they are equal under the eyes of Python by using `==`.

```python
(True == 1) & (False == 0)
```

We can even do operations using them!

```python
True**2 - 2*True + False
```

#### Lists
The last object for the day is the list object. These work the way you probably would like lists to work. We can collect any other objects into a sequence for safe-keeping.

```python
[1, 
 2,  
 3,     
 4] # An example of a list
```

Let's assign our first variable to some list and see what operations we can do with lists.

```python
contlist = [3.14, 
            'i am in a list', 
            None, 
            True, 
            False, 'why did pi get to be first']
```

By writing `contlist = `, we are saying "anytime `contlist` is written, think of the list it represents". This allows us to do operations quicker. For example, if we wanted to add an element, we can use `list_name.append(element_to_add)`.

```python
contlist.append(44)
contlist
```

We can also add multiple elements using `list_name.extend([item1, item2,...])`.

```python
contlist.extend([45,46,47])
contlist
```

With lists, indexing begins at 0. That is, the first item in `contlist` is the 0-th item. To find that item, we use `list_name[index_of_item]`.

```python
contlist[0]
```

To select a sequence of items, we can use `list_name[start_index:stop_index]`.

```python
print(contlist[0:3])
print(contlist[3:]) # leaving stop index empty means end at the end of list
print(contlist[:3]) # leaving start index empty means start at start of list
```

To remove items from lists, we don't want to keep scrolling and editing the list in the cell it exists in. Instead, we want to use `list_name.pop(item_index)` or `list_name.remove(item)`.

```python
contlist.pop(0)
contlist
```

```python
contlist.insert(0, 3.14) # insert 3.14 at 0-th index
contlist.remove(None)
contlist
```

For a list of floats, we can compute the maximum and minimum as `max(list)` and `min(list)` respectively.

```python
min([1,2,3,4])
```

#### For Loops
One canonical method in Python coding is using a "for loop". When we need to something iteratively, this may be the first starting point. We will discuss these today and tomorrow. Lets learn through practice.

```python
contlist = []        # start with an empty list
for i in range(10):  # for each number from 0 to 9
    contlist.append(i**2) # append the number squared to the list
    
    
contlist
```

These are incredibly powerful tools. But, can become unruly so be careful!

```python
contlist = []
for i in range(10000000):
    for j in range(10000):
        contlist.append(i*j**2)
```

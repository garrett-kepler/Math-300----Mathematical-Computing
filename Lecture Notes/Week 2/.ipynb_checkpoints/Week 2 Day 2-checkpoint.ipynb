### Numpy
Numpy is one of the most crucial libraries when doing mathematical computing. The reason numpy is so useful is its computational efficiency in most regards.

Numpy introduces a different type of object to the list of Python objects we know now: the numpy array. Similar to a list, a numpy array is a collection of objects. However, unlike a list, applying a function to a numpy array applies the function immediately to the elements.

```python
contlist = [1, 1, 4, 5, 2, 6] # a list
contarray = np.array(contlist) # the numpy array version
type(contlist), type(contarray)
```

```python
contlist[0]
```

The only restriction on numpy array elements are that they are all the <u>same type of object</u>. If not, then numpy will often make them the same type.

```python
np.array([1, '1'])
```

So, how do we work with numpy arrays? The way we would work with numbers (not lists).

```python
4 * contlist + [1]
```

```python
contarray
```

```python
4 * contarray + 1
```

We also get a speed up when inputting functions:

```python
n = 100000
x = np.linspace(0, 2*np.pi, n)
```

```python
x
```

```python
# using list indexing
%timeit y = [math.sin(x[i]) + math.exp(x[i]**2) for i in range(n)]

# using numpy ufuncs
%timeit y = np.sin(x) + np.exp(x**2)
```

The reason for this speed up is called <i> vectorization </i>. Numpy functions that act on arrays (called <i> universal functions </i> or <i> ufuncs</i>) are designed to execute a single instruction on an entire data set using hardware instructions giving a dramatic speed up in most scenarios.

We can design functions so that they benefit from this same speed up as well. That is, we can <i>vectorize</i> a function:

```python
def f(n):
    return math.sin(n) + math.exp(n**2)
```

```python
x
```

```python
# method 1


vectf = map(f, x) # for each element in a vector x, apply f
```

```python
# method 2


g = np.vectorize(f)
```

```python
%timeit out = list(map(f,x))

%timeit out = g(x)

%timeit out = np.sin(x) + np.exp(x**2)
```

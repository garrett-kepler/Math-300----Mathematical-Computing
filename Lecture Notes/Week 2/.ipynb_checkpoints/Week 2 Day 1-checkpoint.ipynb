### Algorithmic Complexity
Firstly, let us discuss what we mean when code is <i>complex</i>. There is often a naive way to solve a problem in coding, we say <i> brute-force</i>. It gets the job done, but it probably takes a while to run as we increase the size of our input. To describe the difference between brute-force code and <i>optimized</i> code, we will use "Big-O Notation".

Consider scanning through a list to see if an element is in the list:

```python
def scan(somelist, someitem):
    for i in somelist:
        if i == someitem:
            return True
    return False
```

Of course, the best case scenario (meaning costing the least number of operations) is that `someitem` is just the first item in `somelist`. Then, we would only have to perform one operation. The worst case scenario (meaning costing the most number of operations) is that `someitem` is not in the list. In this case, if `somelist` is of length $n$, then we have to perform $n$ operations in the worst case. So, we would say `scan()` is a $\mathcal{O}(n)$ algorithm (of order $n$).

Consider another version:

```python
def scan_really_badly(somelist, someitem):
    for i in somelist:
            for j in somelist:
                    if i == someitem:
                        return True
    return False
```

In `scan_really_badly()`, the worst case scenario is that for each item of the $n$ items in `somelist`, we iterate through all $n$ items of `somelist`. So, in the worst case, we have to do $n^2$ operations. Thus, we would say `scan_really_badly()` is a $\mathcal{O}(n^2)$ algorithm (of order $n^2$).

Clearly, in the worst case, `scan()` will always run faster than `scan_really_badly()` which we can test.

```python
import time 
```

```python
n = 1000000000
somelist = (0 for i in range(n))
someitem = 1


start = time.time()
scan( somelist , someitem)
stop = time.time()
print("scan() took "+str(stop-start)+" seconds")

start = time.time()
scan_really_badly( somelist , someitem)
stop = time.time()
print("scan_really_badly() took "+str(stop-start)+" seconds")
```

This is what in essence big-O notation captures: given an input of size $n$, how many operations occur in the worst case? The more that must occur, the more complex the algorithm is.

For reference, here's some complexities of common functions we use:

   - `seq[i]` returns the i-th item of the sequence. Complexity: $\mathcal{O}(1)$.
   - `seq[i:j]` returns the $k$ items between `seq[i]` and `seq[j]`. Complexity: $\mathcal{O}(k)$.
   - `len(seq)` returns the length of the sequence. Complexity: $\mathcal{O}(1)$.
   - `for item in list:` iterates through a list. Complexity: $\mathcal{O}(n)$.
   - `list.pop(-1)` removes the last object in a list. Complexity: $\mathcal{O}(1)$.
   - `list.pop(0)` removes the first object in a list. Complexity: $\mathcal{O}(n)$.
   
Note: big-O notation is very nuanced. For more formality, please see https://en.wikipedia.org/wiki/Big_O_notation. 

In general, we care only about the highest power of $n$ in our complexity. That is, $\mathcal{O}(1+6n+n^2)$ is not correcly formatted unless written as $\mathcal{O}(n^2)$.

### Overflow & Underflow
You can get Python to perform operations on very small and very large floats only to a ceratin digit count. Beyond this point, Python tells you there is nothing or there is an error.

```python
import sys
print(sys.float_info.min)
```

Up to 308 zeros, we are fine!

```python
10.0**(308) 
```

Any point afterward, we have <i>overflow</i>:

```python
10.0**(400)
```

Likewise, down to 308 decimals we are fine:

```python
print(sys.float_info.min)
```

```python
10.0**(-308)
```

After this point, we will receive something strange. We call this <i>underflow</i>:

```python
10.0**(-400)
```

While this may not be <i>necessarily</i> be zero. Practically, it is close enough. This leads to issues in code as well:

```python
1 / 10.0**(-400)
```

To work around this, we typically use the natural logarithm. Recall that for $x\in \mathbb{R}$,

$\begin{align*}
x&=e^{\log(x)}
\end{align*}$

Thus, to calculate `10.0**(400)` and `10.0**(-400)` accurately, we can use a very important library in Python: numpy.

```python
import numpy as np
```

```python
numbers = [1000, 2000, 5000, 8000, 12000,15000]

log_product = 0

for x in numbers:
    log_product += np.log(x)

print("Log of the product:", log_product)
```

```python
np.e**(log_product)
```

```python
np.mean(numbers)
```

### Guess & Checking
Brute-forcing using for-loops often results in a lot of guess and checking. Consider using the `while` statement instead:

```python
x = 0
while x < 10: # until this condition is met, iterate
    x += 1
    print(x)
```

The `while` statement will iterate until a certain condition is met. This can mean the cell will never cease, so be careful!


Another useful solution to guess and checking is <i>recursion</i>. Similar to the mathematical meaning of recursion, recursion in programming simply means using the same function within its own definition. 

A great example: computing the $n!=n\cdot (n-1)\cdots 3\cdot 2\cdot 1$.

```python
def factorial(n):
    if n <= 1: # define a base case
        return 1
    else:      # define the iteration case
        return n * factorial(n - 1)

print(factorial(4))
```

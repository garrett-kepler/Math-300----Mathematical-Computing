Today, we'll gain more experience using pandas to read and write datasets.

The main object in pandas is the <i>pandas dataframe</i> which is similar to a spreadsheet or table. Moreover, we can convert the dataframe into whatever format we like.

For example, consider the following table.

| Name | Age | Major |
|------|----:|--------|
| Alice | 20 | Mathematics |
| Bob | 22 | Physics |
| Charlie | 19 | Biology |

If I wanted to convert this into a dataframe, I could do the following:

```python
import pandas as pd
```

```python
Name = ['Alice', 'Bob', 'Charlie']
Age = [20, 22, 19]
Major = ['Mathematics', 'Physics', 'Biology']
```

```python
dataframe = pd.DataFrame({
    'Name': Name,
    'Age': Age,
    'Major': Major
})
```

```python
dataframe
```

```python
type(dataframe)
```

```python
dataframe.info() # basic information
```

```python
dataframe.describe() # summary statistics
```
